```python
import sys
!{sys.executable} -m pip install pycryptodome

from Crypto.Cipher import DES
from Crypto.Util.Padding import pad, unpad
from Crypto.Random import get_random_bytes
import binascii

class DESExample:
    def __init__(self):
        self.raw_key = self.generate_symmetric_key()
        self.input_message = input("Enter message to encrypt: ")

        # Encrypt the input message
        encrypted_data = self.encrypt(self.raw_key, self.input_message.encode())
        print("Encrypted message (hex):", binascii.hexlify(encrypted_data).decode())

        # Decrypt the message
        decrypted_message = self.decrypt(self.raw_key, encrypted_data)
        print("Decrypted message:", decrypted_message.decode())


    def generate_symmetric_key(self):
        # Generate a random key for DES (must be 8 bytes)
        key = get_random_bytes(8)
        print("DES Symmetric key (hex):", binascii.hexlify(key).decode())
        return key

    def encrypt(self, key, clear):
        # Create a DES cipher object
        cipher = DES.new(key, DES.MODE_CBC)
        # Pad the plaintext
        padded_clear = pad(clear, DES.block_size)
        # Encrypt the padded plaintext
        iv = cipher.iv
        encrypted = iv + cipher.encrypt(padded_clear)  # Prepend IV to ciphertext
        return encrypted
    def decrypt(self, key, encrypted):
        # Extract the IV from the beginning of the encrypted data
        iv = encrypted[:DES.block_size]
        ciphertext = encrypted[DES.block_size:]
        # Create a DES cipher object with the same IV
        cipher = DES.new(key, DES.MODE_CBC, iv)
        # Decrypt the ciphertext
        decrypted_padded_clear = cipher.decrypt(ciphertext)
        # Unpad the decrypted plaintext
        return unpad(decrypted_padded_clear, DES.block_size)

if __name__ == "__main__":
    DESExample()
```

```python
def power(a, b, mod):
    """Calculates (a^b) % mod using recursive exponentiation by squaring."""
    if b == 1:
        return a % mod
    t = power(a, b // 2, mod)
    if b % 2 == 0:
        return (t * t) % mod
    else:
        return ((t * t) % mod * a) % mod


def main():
    # Input values
    n = int(input("Enter the value of n (prime number): "))
    g = int(input("Enter the value of g (generator): "))
    x = int(input("Enter the value of x for the first person: "))
    a = power(g, x, n)
    y = int(input("Enter the value of y for the second person: "))
    b = power(g, y, n)

    # Calculate the shared keys
    key1 = power(b, x, n)  # Key for the first person
    key2 = power(a, y, n)  # Key for the second person
    print(f"Key for the first person is: {key1}")
    print(f"Key for the second person is: {key2}")

if __name__ == "__main__":
    main()
```

```python
import hashlib
def sha1():
    try:
        md = hashlib.sha1()
        print("Message digest object info: ")
        print(f" Algorithm = {md.name}")
        print(f" Provider = {hashlib.__name__}")
        print(f" ToString = {md}")

        input_str = ""
        md.update(input_str.encode())
        output = md.digest()
        print()
        print(f"SHA1(\"{input_str}\") = {bytes_to_hex(output)}")

        input_str = "abc"
        md = hashlib.sha1()
        md.update(input_str.encode())
        output = md.digest()
        print()
        print(f"SHA1(\"{input_str}\") = {bytes_to_hex(output)}")
        input_str = "abcdefghijklmnopqrstuvwxyz"
        md = hashlib.sha1()
        md.update(input_str.encode())
        output = md.digest()
        print()
        print(f"SHA1(\"{input_str}\") = {bytes_to_hex(output)}")
        print()
    except Exception as e:
        print(f"Exception: {e}")
def bytes_to_hex(b):
    return ''.join([f'{byte:02X}' for byte in b])
if __name__ == "__main__":
    sha1()

```
